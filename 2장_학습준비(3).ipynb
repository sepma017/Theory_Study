### 행렬 곱
- 행렬 곱은 각 원소끼리의 곱셈보다 조금 더 복잡한 연산을 의미

- 행렬 곱은 앞쪽 행렬의 각 행의 원소와 뒤쪽 행렬의 각 열의 원소를 곱한 후 모든 값을 더해 새로운 행렬의 원소를 만든다.

- 시그마(Σ)는 행렬 곱에서 각 원소를 곱한 값의 합계를 계산할 때 큰 도움이 됨.
- 행렬 곱은 특정한 조건을 제외하고는 곱하는 두 행렬의 순서를 바꿀 수 없다.
- 행렬의 곱은 앞쪽 행렬의 열 수와 뒤쪽 행렬의 행 수가 일치해야 한다.

코딩테스트_2-23.png

- l x m 행렬과 m x n 행렬의 곱은 l x n 행렬이 된다.

코딩테스트_2-24.png

- 행렬의 곱을 dot 함수를 이용하면 결괏값을 간단히 구할 수 있다.

```python
# dot 함수로 행렬 곱 계산
import numpy as np

a = np.array([[0, 1, 2], [1, 2, 3]])
b = np.array([[2, 1], [2, 1], [2, 1]])

print(np.dot(a, b))
```

- a의 열 수와 b의 행 수가 일치하지 않으면 넘파이는 오류 표시함.
- 행렬 곱은 대량의 데이터를 한 번에 빠르게 처리할 수 있으며 딥러닝을 신속하게 수행하기 위해서도 중요한 연산임.

### 행렬 전치
- **전치** : 행렬에서 행과 열이 바뀜.
- 넘파이에서는 행렬 이름 뒤에 .T를 붙이면 전치가 됨.

코딩테스트_2-25.png

```python
# 전치
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])
print(a.T)
```

- 행렬 곱은 기본적으로 앞쪽 행렬의 열 수와 뒤쪽 행렬의 행 수가 일치해야 함.
- 그런데 일치시키지 않아도 전치시키면 행렬 곱이 가능한 경우도 있음.

- 정규분포를 따르는 데이터의 분포는 넘파이를 이용해 간단히 생성할 수 있다.
- 넘파이의 random.normal 함수로 정규분포를 따르는 난수를 생성하고 맷플롯립(Matplotlib)의 hist함수로 히스토그램을 표현하는 예

```python
# 정규분포 히스토그램
import numpy as np
import matplotlib.pyplot as plt

# 정규분포를 따르는 난수 생성(평균 50, 표준편차 10, 10000개 생성)
x = np.random.normal(50, 10, 10000)

# 히스토그램
plt.hist(x, bins=50)        # 50은 막대의 수
plt.show()
```
