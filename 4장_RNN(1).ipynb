# RNN

 - 시계열 데이터를 다루는 신경망의 한 종류인 RNN(순환 신경망)을 살펴본다.
 - RNN 전체 구조를 살펴본 후, 수식과 코드를 활용해 간단한 구조의 RNN을 구현한다.
 - 파이썬 코드를 통해 RNN의 작동 방식을 파악한다.

## RNN 개요
- **RNN**(Recurrent Neural Network, 순환 신경망) : 은닉층이 순환하는 구조

이론_3-54.png

- 은닉층의 출력은 다음번 입력과 함께 은닉층의 입력이 된다. (은닉층이 순환하는 구조) => **재귀(recurrence)**(자기 자신을 순환하는 것)
- RNN의 은닉층은 이전 시점의 은닉층으로부터 영향을 받기 때문에 신경망 자체도 이전 시점의 영향을 받게 된다. (RNN은 과거의 기억을 이용해 판단할 수 있다.)
- RNN은 자연어처럼 매번 입력값의 길이가 다른 데이터를 다룰 수 있다.
- RNN은 음성, 문장, 동영상 등 시간에 따라 변하는 데이터, 즉 시계열 데이터를 다룰 때 적용한다. (시계열 데이터가 RNN의 입력값)
- 은닉층이 연속적인 시간상에서 모두 연결되어 있어 어떤 의미에서 심층 신경망을 알 수 있다. 
- RNN은 역전파로 학습할 수 있지만, 오차의 계산 방법이 일반적인 신경망과는 다르다.
- RNN에서는 오차가 과거로 거슬러 올라가지만, 특정 시점에서 오차의 출력 기울기는 출력층에서 역전파해 온 출력값의 기울기와 이후 시점에서 거슬러 온 출력 값 기울기의 합이 된다. 
- 이런 방식으로 전체 시간에 걸쳐 오차를 역전파시켜 기울기를 계산하고 가중치와 편향을 수정한다.

이론_3-55.png

- RNN은 시계열 형태의 심층 신경망 구조 => 몇 개의 층에 걸쳐 오차를 전파시키면 기울기가 소실 또는 발산하는 문제가 있다. 이전 시점에서 이어받은 데이터에 반복해서 같은 가중치를 곱하기 때문에 일반적인 신경망에 비해 이와 같은 문제가 더 심해진다.(일반적인 신경망은 순환 없이 층마다 가중치가 다르므로 RNN에 비해 이런 문제가 발생하기 어렵다.)
- **장기 의존성 문제(problem of Long-Term Dependency)**: 데이터의 장기 기억에 관한 문제(RNN층으로는 단기 기억은 유지 되지만 장기 기억을 유지하기 어렵다.)
- 예측 정확도를 향상시키려면 장기 의존성에 맞는 신경망 필요 => **LSTM(Long Short Term Memory)** , **GRU(Gated Recurrent Unit)** 등 게이트 구조를 도입한 RNN이 제안됨.
